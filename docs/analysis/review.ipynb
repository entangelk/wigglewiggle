```python
df_review = pd.read_csv('../data/wiggle.photo_review.csv')
```

```python
df_temp = pd.concat([df_review['item_name'],df_review['review_count']],axis=1)
```

```python
df_temp = df_temp.drop_duplicates().sort_values('review_count',ascending=False)
```

```python
plt.figure(figsize=(15, 6))

sns.barplot(data=df_temp[:10], x='item_name', y='review_count')

plt.xticks(rotation=45)
plt.title('리뷰 수 top 10')
plt.show()
```
